```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sinasc = pd.read_csv('SINASC_RO_2019.csv')
sinasc.columns
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
plt.figure(figsize = (4, 6))
sns.boxplot(data = sinasc, y = 'IDADEMAE', color = 'skyblue')
plt.title('Boxplot Idade da Mãe')
plt.ylabel('Idade da Mãe')
plt.grid(True, axis = 'y', linestyle = '--', alpha = 0.7)
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
plt.figure(figsize = (4, 6))
sns.boxplot(data = sinasc, y = 'PESO', color = 'lightpink')
plt.title('Boxplot Peso do bebê')
plt.ylabel('Peso do bebê')
plt.grid(True, axis = 'y', linestyle = '--', alpha = 0.7)
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
plt.figure(figsize = (8, 4))
sns.displot(data = sinasc, color = 'skyblue', x = 'IDADEMAE', bins = 20, element = 'step')
plt.title('Hist. Idade da Mãe')
plt.xlabel('Idade da Mãe')
plt.ylabel('Frequência')
plt.grid(True, axis = 'x', linestyle = '--', alpha = 0.25)
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
plt.figure(figsize = (8, 4))
sns.displot(data = sinasc, color = 'lightpink', x = 'PESO', bins = 20, element = 'step')
plt.title('Hist. Peso do bebê')
plt.xlabel('Peso do bebê')
plt.ylabel('Frequência')
plt.grid(True, axis = 'x', linestyle = '--', alpha = 0.25)
plt.show()
```

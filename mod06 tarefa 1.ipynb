```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sinasc = pd.read_csv('SINASC_RO_2019.csv')
sinasc.columns
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
variaveis = ['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE']

#gráficos
plt.figure(figsize=(16, 12))
for i, var in enumerate(variaveis):
    plt.subplot(3, 2, i + 1)
    sns.countplot(data = sinasc, x = var, hue = var, palette = 'pastel', legend = False, 
                  order = sinasc[var].value_counts().index)
    plt.title(f'Distribuição de {var}')
    plt.xlabel(var)
    plt.ylabel('Frequência')
    plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
media_apgarc = sinasc.groupby('CONSULTAS')['APGAR5'].mean().reset_index()
media_apgarc = media_apgarc.sort_values(by = 'CONSULTAS')

#gráfico
plt.figure(figsize = (10, 6))
sns.barplot(data = media_apgarc, x = 'CONSULTAS', y = 'APGAR5', hue = 'CONSULTAS',
            palette = 'pastel', legend = False)
plt.title('Média de APGAR5 por número de consultas')
plt.xlabel('Número de consultas')
plt.ylabel('média do APGAR5')
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
variaveis = ['CONSULTAS', 'SEXO', 'GESTACAO', 'ESTCIVMAE']

#gráficos
plt.figure(figsize=(16, 12))

for i, var in enumerate(variaveis):
    plt.subplot(2, 2, i + 1)
    sns.countplot(data=sinasc, x=var, hue=var, palette='pastel', legend=False,
                  order=sinasc[var].value_counts().index)
    plt.title(f'Distribuição de {var}')
    plt.xlabel(var)
    plt.ylabel('Frequência')
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
